# 1. Standardization

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
#Generrate 1000 noramal random intrgers with specified mean and std.
draw=norm.ppf(np.random.random(100000),loc=1000,scale=200).astype(int)
sns.distplot(draw)                      # loc=mean and scale=std.
```

```python
z=[]
for i in draw:
    z.append((i-1000)/200 )
```

```python
a=np.array(z)
```

```python
print(a.mean())
print(np.round(a.mean()))
# whenever we standertize or standard normal converted mean becomes zero s yo can see
```

```python
print(a.std())
print(np.round(a.std()))
```

```python
sns.distplot(a)
```
