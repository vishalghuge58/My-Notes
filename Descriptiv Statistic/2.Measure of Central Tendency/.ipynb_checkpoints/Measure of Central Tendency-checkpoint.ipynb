```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_data=pd.read_csv('Iris.csv')
iris_data.drop('Id',inplace=True,axis=1)
iris_data.head()
```

```python
iris_data.rename(columns={'SepalLengthCm':'sl','SepalWidthCm':'sw','PetalLengthCm':'pl','PetalWidthCm':'pw'},inplace=True)
iris_data.tail()
```

```python
iris_data['Species'].value_counts()
```

```python
iris_setosa=iris_data[iris_data['Species']=='Iris-setosa']
iris_versicolor=iris_data[iris_data['Species']=='Iris-versicolor']
iris_virginica=iris_data[iris_data['Species']=='Iris-virginica']
```

# 1. Mean

```python
## SepalWidth

print(np.mean(iris_setosa['sw']))
print(np.mean(iris_versicolor['sw']))
print(np.mean(iris_virginica['sw']))
```

```python
## SepalLength

print(np.mean(iris_setosa['sl']))
print(np.mean(iris_versicolor['sl']))
print(np.mean(iris_virginica['sl']))
```

```python
## Petalwidth

print(np.mean(iris_setosa['pw']))
print(np.mean(iris_versicolor['pw']))
print(np.mean(iris_virginica['pw']))
```

```python
## PetalLength

print(np.mean(iris_setosa['pl']))
print(np.mean(iris_versicolor['pl']))
print(np.mean(iris_virginica['pl']))
```

## The problem of Outliers

```python
#The range of setosa petal length is between 1 to 2

plt.hist(iris_setosa['pl'])
```

```python
# If we add the Outlier of 50 at the end
np.mean(np.append(iris_setosa['pl'],50))
```

```python
# The Solution is Median
```

## 2. Median

```python
## PetalLength

print(np.median(iris_setosa['pl']))
print(np.median(iris_versicolor['pl']))
print(np.median(iris_virginica['pl']))
```

### Good Thing about Median

```python
np.median(np.append(iris_setosa['pl'],50))
```

```python
# the good thing about the median is until and unless more than half of the data becomes outlier its not affected

# Ex.,
arry=np.array([1,4,5,7,1000,5000,10000])
np.median(arry)

# as you can see here three values are outlier but still not affected
```

## 3.Mode

```python
from scipy import stats

print(stats.mode(iris_setosa['pl']))
print(stats.mode(iris_versicolor['pl']))
print(stats.mode(iris_virginica['pl']))
```
